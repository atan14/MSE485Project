```python
basis_points = [[0.0, 0.0, 0.0], #0
                [0.5, 0.5, 0.5], #1
                [0.25, 0.75, 0.75], #2
                [0.25, 0.25, 0.75],#3
                [0.5, 0.0, 0.5],#4
                [0.25, 0.75, 0.25],#5
                [0.0, 0.5, 0.5],#6
                [0.75, 0.75, 0.75],#7
                [0.5, 0.5, 0.0], #8
                [0.75, 0.25, 0.25],#9
                [0.75, 0.25, 0.75],#10
                [0.0, 0.0, 0.5],#11
                [0.75, 0.75, 0.25], #12
                [0.5, 0.0, 0.0], #13
                [0.25, 0.25, 0.25], #14
                [0.0, 0.5, 0.0]] #15
```

```python
basis1 = basis_points[1]
basis3 = basis_points[3]
basis5 = basis_points[5]
basis7 = basis_points[7]
basis9 = basis_points[9]
basis11 = basis_points[11]
basis13 = basis_points[13]
basis15 = basis_points[15]
```

```python
basis_set = [basis3, basis5, basis7, basis9, basis1, basis11, basis13, basis15]
basis_dir1 = [basis_set[i] for i in [4, 5, 6, 7]]
basis_dir2 = [basis_set[i] for i in [0, 1, 2, 3]]
# [0, 1, 2, 3] --> direction2
# [4, 5, 6, 7] --> direction1
```

```python
direction1 = [[0.25, 0.25, 0.25],
             [-0.25, -0.25, 0.25],
             [-0.25, 0.25, -0.25],
             [0.25, -0.25, -0.25]]
direction2 = [[-0.25, -0.25, -0.25],
             [-0.25, 0.25, 0.25],
             [0.25, -0.25, 0.25],
             [0.25, 0.25, -0.25]]
```

```python
from operator import add, sub, mod
for idx, basis in enumerate(basis_set):
    if idx in [0, 1, 2, 3]:
        for dir_i in direction2:
            loc1 = list(map(add, basis, dir_i))
            basis_mapping = list(map(mod, loc1, [1, 1, 1]))
            if basis_mapping in basis_dir1:
                for dir_j in direction1:
                    loc2 = list(map(add, loc1, dir_j))
                    print (basis, "--->", loc1, "--->", loc2)
#                     print (basis, "--(", dir_i, ")-->", loc1, "--(", dir_j, ")-->", loc2)
            elif basis_mapping in basis_dir2:
                for dir_j in direction1:
                    loc2 = list(map(add, loc1, dir_j))
                    print (basis, "--->", loc1, "--->", loc2)
#                     print (basis, "--(", dir_i, ")-->", loc1, "--(", dir_j, ")-->", loc2)
    else:
        for direction in direction1:
            loc1 = list(map(add, basis, direction))
            basis_mapping = list(map(mod, loc1, [1, 1, 1]))
            if basis_mapping in basis_dir1:
                for dir_j in direction1:
                    loc2 = list(map(add, loc1, dir_j))
                    print (basis, "--->", loc1, "--->", loc2)
#                     print (basis, "--(", dir_i, ")-->", loc1, "--(", dir_j, ")-->", loc2)
            elif basis_mapping in basis_dir2:
                for dir_j in direction1:
                    loc2 = list(map(add, loc1, dir_j))
                    print (basis, "--->", loc1, "--->", loc2)
#                     print (basis, "--(", dir_i, ")-->", loc1, "--(", dir_j, ")-->", loc2)
    print ()
```
